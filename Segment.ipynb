```python
#scraping an urdu website to make a dictionary from

import requests
from bs4 import BeautifulSoup

#jang newspaper website
url = 'https://jang.com.pk/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser') #parsing the page
#getting text from page
text = soup.get_text()
#array to store urdu words
urdu_words = []

#loop to see if any word has urdu characters and storing it in urdu words array
for word in text.split():
    if any(ord(c) > 127 for c in word):
        urdu_words.append(word)
        
#print(urdu_words)
urdu_words
```

```python
import numpy as np
import pandas as pd

# storing all urdu words in this file in urdu words 2 array
urdu_words2 = [i for i in np.load('corpus.npy')]
urdu_words2
```

```python
#reading all urdu words from this file and storing them to store more urdu words
#this txt file taken from github

with open('all-words.txt', 'rt', encoding="utf-8") as f:
    reader = csv.reader(f)
    urdu_words3 = list(reader)
f.close()

urdu_words3
```

```python
# Merging all 3 dictionaries to make one urdu dictionary with a larger number of words

Urdu_words = urdu_words.copy()

for word in urdu_words2:
    # to not add any duplicates in the dictionary
    if word not in Urdu_words:
        Urdu_words.append(word)

#merging third array with previous two
for word in urdu_words3:
    if word not in Urdu_words:
        Urdu_words.append(word)

Urdu_words
```

```python
#function to segment words 

#The function splits an input text into individual words based on the longest possible match from a dictionary of urdu words
def word_segment(text, dictionary):
    start = 0
    size = len(text)
    words = []  
    #loops throuhh start till end of text
    while start < size:
        max_word = text[start]
        #checks every possible consecutive substring in the text starting from the current starting position
        for i in range(1, size - start + 1):
            if text[start:start + i] in dictionary:
                #if the substring exists in the dictionary, it sets it as the current max_word
                max_word = text[start:start + i]
                
        words.append(max_word)
        start += len(max_word)
        
    return words
```

#### Sentence Segmentation

```python
#web scraping to store words that usually come at the end of a sentence in an array

import requests
from bs4 import BeautifulSoup
import re

url = 'https://jang.com.pk/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
text = soup.get_text()

words = []
# check for words that come before ?, !, ., or -
words = re.findall(r'\b[\u0600-\u06FF]+(?=[\s?!.-])', text)

print(words)
```

```python
#function to segment sentences
def sentence_segment(text, stop):#stop has words that are stopping criteria 
    sentences = []
    sentence_words = []
    for word in text.split():
        #if word[-1] in ['.', '?', '!', '-']:#checks for these punctuation which comes at end of sentences 
        #breaks into too many sentences so not used
        sentence_words.append(word)

        if len(sentence_words) > 5: # so text not broken into many small parts
            #if number of words exceed 20 or the word is found in stopping criteria array, add fullstop and append to sentences 
            if len(sentence_words) > 20 or word in stop: 
                sentence = ' '.join(sentence_words) + '۔'
                sentences.append(sentence)
                sentence_words = []
    
 #To check if there are any words remaining in the sentence_words list after going through all words in text
    if len(sentence_words) > 0:
        #append remaing words as a sentence in the sentence array
        sentence = ' '.join(sentence_words) 
        sentences.append(sentence)
        
    return sentences
```

### Function for word segmentation:
#### The function takes an input text and a dictionary of Urdu words as parameters.
#### It enters a while loop that continues until the end of the text is reached.
#### Inside the while loop, the function considers every possible consecutive substring in the text starting from the current start position.
#### For each substring, the function checks if it exists in the given dictionary. If it does, then the substring is considered as a possible word.
#### The function then chooses the longest possible match from the dictionary as the current word.
#### After the loop ends, the current word is appended to a list of words and the starting position is incremented by the length of the word.
#### Finally, the function returns the list of words.

### Function for sentence segmentation:
#### The function takes an input text and an array of words that will pose as stopping criteria, as parameters.
#### It initializes two empty lists: one to hold the segmented sentences and another to hold the words of the current sentence being constructed.
#### The function enters a for loop that iterates over each word in the given text.
#### For each word, the function appends it to the list of sentence words.
#### The function checks if the length of the sentence_words list is greater than 5.
#### If the length is greater than 20 or the current word is found in the stop array, the function adds a full stop to the end of the current sentence, joins the sentence words together, appends the resulting sentence to the sentences list, and then clears the sentence_words list.
#### After the for loop ends, the function checks if there are any words remaining in the sentence_words list. If there are, the function joins the remaining words together and appends the resulting sentence to the sentences list.
#### Finally, the function returns the list of segmented sentences.
